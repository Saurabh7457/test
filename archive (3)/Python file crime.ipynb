#### EXTRACTING JSON FILE 

```python
import requests

url = "https://data.cityofchicago.org/resource/4aki-r3np.json"

response = requests.get(url)
print(response.json())
```

```python
response.json()
```

#### EXTRACTING DATA WITH SPECIFIC COLUMNS

```python
import pandas as pd
```

```python
df=pd.DataFrame(response.json()[0:1000])[['ssl_score','age_curr','idoc_res_zip_code','sex_code_cd','race_code_cd', 'majority_dist','latest_dist_res']]
df
```

#### DATA CLEANING

```python
df.isna().sum()
```

```python
df=df.fillna('De-identified')
df
```

```python
df.isna().sum()
```

#### CONVERTING TO CSV

```python
df.to_csv('Criminal Profile.csv')
```
